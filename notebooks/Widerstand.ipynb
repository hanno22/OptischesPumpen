```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data =  np.loadtxt('../data/WiderstandsMessung.csv', skiprows=1, delimiter=',')
U = data[:,0]
I = data[:,1]

slope, intercept, r, p, std_err = stats.linregress(I, U)


fig, ax = plt.subplots()
ax.set(xlabel='I / A', ylabel='U / V')
ax.scatter(I, U)
ax.plot(I, slope * I + intercept)
```
